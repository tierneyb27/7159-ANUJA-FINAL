# Subset-Sum Problem (SSP)

The **Subset Sum problem (SSP)** is defined as follows:

>Given a set $S = \{x_1, x_2, \ldots, x_n\}$ of positive integers,
>and a positive integers $t$, is there a subset of $S$ whose sum is equal to $t$?

This is the **decision version** of SSP as it only asks for a true/false answer.

#### Example

$S=\{1,2,3,10\}$ and $t=13$.

A solution is given by $\{1,2,10\}$. There is also another solution: $\{3,10\}$.

## Exhaustive search

An **exhaustive search algorithm** for this problem can written *iteratively* as follows:

**INPUT:** A set $S = \{x_1, x_2, \ldots, x_n\}$ of positive integers, and a positive integer $t$.

**OUTPUT:** $(c_1,\ldots,c_n)$ such that $\sum_{i=1}^n c_i x_i = t$

1. **for all** $(c_1,\ldots,c_n)\in\{0,1\}^n$ **do**
2. $\quad$ **if** $\sum_{i=1}^n c_i x_i = t$ **then**
3. $\qquad$ **return** $(c_1,\ldots,c_n)$

# **Recursive version of the exhaustive search algorithm.**
```
RECURSIVE-SSP(S, t, i)
INPUT: A set S = {x₁, x₂, ..., xₙ} of positive integers, target sum t, current index i  
OUTPUT: (c₁,...,cₙ) such that Σcᵢxᵢ = t, or NULL if no solution exists

1. if t == 0 then
2.     return (0,0,...,0)
3. if i > n or t < 0 then  
4.     return NULL
5. result₁ = RECURSIVE-SSP(S, t - S[i], i + 1)
6. if result₁ ≠ NULL then
7.     result₁[i] = 1
8.     return result₁
9. result₂ = RECURSIVE-SSP(S, t, i + 1)
10. if result₂ ≠ NULL then
11.    result₂[i] = 0  
12.    return result₂
13. return NULL

Initial call: RECURSIVE-SSP(S, t, 1)
```

## Dynamic Programming.

The above iterative pseudocode considers the subsets of $S$ as the search space.
Another way to search for solutions is to iteratively build the answer for smaller target values $t' = 0, 1, 2, 3, \ldots$ until we reach $t$.

If we have built all the possible sums from the subset $\{x_1,\ldots,x_k\}$, what other sums become possible if we add $x_{k+1}$ to the set?

# **Answer**

By using xk+1 we can produce some new rules.

To find the new sum's, we can take help from the old sum and then we can add the xk+1 to them. These are the new results are the additional sums we can make.

For Example:

The current possible sums are:
{0,3,5}, and we will add the xk+1 = 4.

Now we can make the new sums by doing:

* 0 + 5 = 5

* 2 + 5 = 7

* 6 + 5 = 11

So, the new list of sums can be said as:

{0, 2, 5, 6, 7, 11}

This idea will help in building a new solution step by step, as like in dynamic programming.

Top to bottom method

````python
def subset_sum_memo(S, t):
    n = len(S)
    memo = {}

    def solve(i, target):
        if target == 0:
            return True
        if i >= n or target < 0:
            return False

        if (i, target) in memo:
            return memo[(i, target)]

        # Thi is the recursive calls
        include = solve(i + 1, target - S[i])
        exclude = solve(i + 1, target)

        # This will store the result in the memo table
        memo[(i, target)] = include or exclude
        return memo[(i, target)]

    return solve(0, t)
````

````python
# Test examples

# Test case 1
S1 = [1, 2, 3, 10]
t1 = 13
print(f"subset_sum_memo({S1}, {t1}) = {subset_sum_memo(S1, t1)}")

# Test case 2
S2 = [2, 4, 6, 8]
t2 = 5
print(f"subset_sum_memo({S2}, {t2}) = {subset_sum_memo(S2, t2)}")

# Test case 3
S3 = [1, 3, 5, 7]
t3 = 8
print(f"subset_sum_memo({S3}, {t3}) = {subset_sum_memo(S3, t3)}")
````

### 2) bottom to top approach

````python
def subset_sum_dp(S, t):
    n = len(S)
    dp = [[False] * (t + 1) for _ in range(n + 1)]

    # Base case
    for i in range(n + 1):
        dp[i][0] = True

    for i in range(1, n + 1):
        for j in range(1, t + 1):
            dp[i][j] = dp[i-1][j]

            if j >= S[i-1]:
                dp[i][j] = dp[i][j] or dp[i-1][j - S[i-1]]

    return dp[n][t]
````

````python
# Test examples

# Test case 1
S1 = [1, 2, 3, 10]
t1 = 13
print(f"subset_sum_dp({S1}, {t1}) = {subset_sum_dp(S1, t1)}")

# Test case 2
S2 = [2, 4, 6, 8]
t2 = 5
print(f"subset_sum_dp({S2}, {t2}) = {subset_sum_dp(S2, t2)}")

# Test case 3
S3 = [1, 3, 5, 7]
t3 = 8
print(f"subset_sum_dp({S3}, {t3}) = {subset_sum_dp(S3, t3)}")
````

# Conclusion



Here, we took into account different techniques or algorithms, to solve the Subset Sum Problem (SSP), which will trace the transition from naive recursive solution to more efficient dynamic programming techniques.

Through utilization of the dynamic programming, we managed to address the huge performance gains with two major startergies. The first method addressed is memoization (top-down), eliminates redundant recursive calls by storing results of previous computations. The secnod method addressed is bottom-up approach, which will build solution to incrementally from small subproblems. The two methods reduce the time complexity to Ο(nxt), where "n" represents the number od elements and "t" represents the target sum.

The key idea here in the dynacmic programming is that whenever we try to insert a new element  xk+1  into the set, achievable sums can be determined by beginning with existing sums and adding  xk+1  to each one of them. This insight allows us to construct solution piecewise, instead of needing to try all possible combinations.

My experimentation and testing confirms that the both approaches produce the same outputs, albeit with significant speed improvements. The bottom up approach technique also precludes the risk of stack overflow inherent with deep recursion in the memoization technique.

Lastly, Subset Sum Problem is an excellent illustration of dynamic programming that transforms a computationally expensive exponential time problem into a far more manageable polynomial-time solution. This optimization is a significant concept in algorithm design and it finds a very broad application to the many problems beyond Subset Sum Problem.
